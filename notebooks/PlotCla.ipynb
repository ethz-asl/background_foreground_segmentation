# Example how to use the dataset

```python
import tensorflow_datasets as tfds
import bfseg.data.meshdist.bfseg_cla_meshdist_labels
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (30,7)

ds_cla = tfds.load("BfsegCLAMeshdistLabels", split="fused")
```

## CLA DATASET

```python
for entry in ds_cla.take(11):
    plt.subplot(1,3,1)
    plt.imshow(entry['image'])
    plt.subplot(1,3,2)
    plt.imshow(entry['label'])
    plt.subplot(1,3,3)
    plt.imshow(entry['image'])
    plt.imshow(entry['label'], alpha = 0.5)
```
