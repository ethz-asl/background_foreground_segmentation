# Example how to use the dataset

```python
import tensorflow_datasets as tfds
import bfseg.data.hive.bfseg_validation_labeled
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (30,7)

ds_arche = tfds.load("BfsegValidationLabeled", split="ARCHE")
ds_cla = tfds.load("BfsegValidationLabeled", split="CLA")
```

## CLA DATASET

```python
for entry in ds_cla.take(1):
    plt.subplot(1,2,1)
    plt.imshow(entry['image'])
    plt.subplot(1,2,2)
    plt.imshow(entry['label'])
```

## ARCHE DATASET

```python
for entry in ds_arche.take(1):
    plt.subplot(1,2,1)
    plt.imshow(entry['image'])
    plt.subplot(1,2,2)
    plt.imshow(entry['label'])
```
