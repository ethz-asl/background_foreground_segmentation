# Example how to use the dataset

```python
import tensorflow_datasets as tfds
import bfseg.data.hive.office_rumlang_validation_labeled
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (30,7)

ds_office = tfds.load("OfficeRumlangValidationLabeled", split="OFFICE")
ds_rumlang = tfds.load("OfficeRumlangValidationLabeled", split="RUMLANG")
```

## OFFICE DATASET

```python
for entry in ds_office.take(1):
    plt.subplot(1,2,1)
    plt.imshow(entry['image'])
    plt.subplot(1,2,2)
    plt.imshow(entry['label'])
```

## RUMLANG DATASET

```python
for entry in ds_rumlang.take(1):
    plt.subplot(1,2,1)
    plt.imshow(entry['image'])
    plt.subplot(1,2,2)
    plt.imshow(entry['label'])
```
