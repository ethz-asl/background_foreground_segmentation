```python
import matplotlib.pyplot as plt
import tensorflow_datasets as tfds
```

```python
from bfseg.utils.datasets import preprocess_bagfile
```

```python
ds_rumlang2 = tfds.load('MeshdistPseudolabels', split='rumlang2', as_supervised=True)
ds_rumlang3 = tfds.load('MeshdistPseudolabels', split='rumlang3', as_supervised=True)
```

```python
ds_rumlang2 = ds_rumlang2.map(preprocess_bagfile)
ds_rumlang3 = ds_rumlang3.map(preprocess_bagfile)
```

```python
for dataset_name, dataset in zip(
    ['ds_rumlang2', 'ds_rumlang3'], [ds_rumlang2, ds_rumlang3]):
    for sample_idx, (image, label, mask) in enumerate(dataset.take(2)):
        fig, ax = plt.subplots(1, 3, constrained_layout=True, dpi=300)
        fig.suptitle(
            f"{dataset_name}, sample no. {sample_idx+1}",
            fontsize=12,
            y=0.75)
        ax[0].imshow(image)
        print(f"image.shape = {image.shape}")
        ax[1].imshow(label)
        print(f"label.shape = {label.shape}")
        ax[2].imshow(mask)
        print(f"mask.shape = {mask.shape}")
```

### Alternative version (just `load_data`)

```python
from bfseg.utils.datasets import load_data
```

```python
batch_size = 2
```

```python
data_rumlang2 = load_data(
    dataset_name="MeshdistPseudolabels",
    scene_type="rumlang2",
    fraction=None,
    batch_size=batch_size,
    shuffle_data=False)
data_rumlang3 = load_data(
    dataset_name="MeshdistPseudolabels",
    scene_type="rumlang3",
    fraction=None,
    batch_size=batch_size,
    shuffle_data=False)
```

```python
for dataset_name, dataset in zip(
    ['data_rumlang2', 'data_rumlang3'], [data_rumlang2, data_rumlang3]):
    for batch in dataset.take(1):
        for sample_idx in range(batch_size):
            image = batch[0][sample_idx]
            label = batch[1][sample_idx]
            mask = batch[2][sample_idx]
            fig, ax = plt.subplots(1, 3, constrained_layout=True, dpi=300)
            fig.suptitle(
                f"{dataset_name}, sample no. {sample_idx+1}",
                fontsize=12,
                y=0.75)
            ax[0].imshow(image)
            print(f"image.shape = {image.shape}")
            ax[1].imshow(label)
            print(f"label.shape = {label.shape}")
            ax[2].imshow(mask)
            print(f"mask.shape = {mask.shape}")
```
